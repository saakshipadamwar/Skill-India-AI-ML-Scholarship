# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
```

```python
olympics = pd.read_csv("summer.csv")
```

### 1. In how many cities Summer Olympics is held so far?

```python
len(olympics["City"].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
gold = []
    
for i in olympics:
    if i == "Medal":
        gold.append(olympics[olympics[i] == "Gold"])

gold = pd.DataFrame(gold[0] , columns = ["Year" ,"City" , "Sport" , "Discipline" , "Athlete" , "Country" , "Gender" , "Event" , "Medal" ])

sports_gold = []

for sport_gold in gold["Sport"].unique():
    sports_gold.append([sport_gold , len(gold[gold["Sport"] == sport_gold])])
    
sports_gold = pd.DataFrame(sports_gold , columns = ["Sport" , "No. of Golds"])
sports_gold.sort_values(by = "No. of Golds" , ascending = False)

print(sports_gold.head().plot(x = "Sport" , y = "No. of Golds" , kind = "bar" , figsize = (10,5)))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
sports = []

for sport in olympics["Sport"].unique():
    sports.append([sport , len(olympics[olympics["Sport"] == sport])])

sports = pd.DataFrame(sports, columns = ["Sport" , "No. of Medals"])
print(sports.sort_values(by = "No. of Medals" , ascending = False).head().plot(x = "Sport" , y = "No. of Medals" , kind = "bar" , figsize = (10,5)))

```

### 4. Which player has won most number of medals? (Top 5)

```python
players = []

for player in olympics["Athlete"].unique():
    players.append([player , len(olympics[olympics["Athlete"] == player])])

players = pd.DataFrame(players, columns = ["Athlete" , "No. of Medals"])
print(players.sort_values(by = "No. of Medals" , ascending = False).head().plot(x = "Athlete" , y = "No. of Medals" , kind = "bar" , figsize = (10,5)))

```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
players_gold = []

for player_gold in gold["Athlete"].unique():
    players_gold.append([player_gold , len(gold[gold["Athlete"] == player_gold])])
    
players_gold = pd.DataFrame(players_gold , columns = ["Athlete" , "No. of Golds"])
print(players_gold.sort_values(by = "No. of Golds" , ascending = False).head().plot(x = "Athlete" , y = "No. of Golds" , kind = "bar" , figsize = (10,5)))

```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
India_gold = []
for i in gold:
    if i == "Country":
        India_gold.append(gold[gold[i] == "IND"])
        
India_gold = pd.DataFrame(India_gold[0] , columns = ["Year" ,"City" , "Sport" , "Discipline" , "Athlete" , "Country" , "Gender" , "Event" , "Medal" ])
print(India_gold.sort_values(by = "Year" , ascending = True).head(1))

```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
events = []

for event in olympics["Event"].unique():
    events.append([event , len(olympics[olympics["Event"] == event])])

events = pd.DataFrame(events, columns = ["Event" , "Frequency"])
print(events.sort_values(by = "Frequency" , ascending = False).head().plot(x = "Event" , y = "Frequency" , kind = "bar" , figsize = (10,5)))

```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
female_gold = []
for i in gold:
    if i == "Gender":
        female_gold.append(gold[gold[i] == "Women"])
        
female_gold = pd.DataFrame(female_gold[0] , columns = ["Year" ,"City" , "Sport" , "Discipline" , "Athlete" , "Country" , "Gender" , "Event" , "Medal" ])

female_sport_gold = []
for sport in female_gold["Sport"].unique():
    female_sport_gold.append([sport , len(female_gold[female_gold["Sport"] == sport])])

female_sport_gold = pd.DataFrame(female_sport_gold , columns = ["Sports" , "No. of females"])
print(female_sport_gold.sort_values(by = "No. of females" , ascending = False).head().plot(x = "Sports" , y = "No. of females" , kind = "bar" , figsize = (10,5)))
```
